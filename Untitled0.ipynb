```python
#  Load styles.csv
df = pd.read_csv("styles.csv", on_bad_lines="skip", engine="python")

print("Full dataframe shape:", df.shape)
df.head()
```

```python
#Filter Women + Boutique Categories

required = [
    "Kurtas", "Tops", "Dresses", "Sarees", "Jeans",  "Tshirts",
    "Leggings","Bras", "Shrugs", "Tunics", "Sleepwear",
    "Innerwear","Jumpsuits","Shirts"
]

women_df = df[
    (df["gender"] == "Women") &
    (df["articleType"].isin(required))
]

print("Filtered women_df shape:", women_df.shape)
print(women_df["articleType"].value_counts())
```

```python
IMG_SIZE = 224
BATCH_SIZE = 16

train_gen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    validation_split=0.2
)

train_data = train_gen.flow_from_directory(
    "dataset",
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='training'
)

val_data = train_gen.flow_from_directory(
    "dataset",
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='validation'
)
```

```python
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense, Dropout, BatchNormalization
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.regularizers import l2
from tensorflow.keras.losses import CategoricalCrossentropy
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
```

```python
from tensorflow.keras.applications import MobileNetV2

base_model = MobileNetV2(
    include_top=False,
    weights="imagenet",
    input_shape=(IMG_SIZE,IMG_SIZE,3)
)

base_model.trainable = False
```

```python
from tensorflow.keras.layers import Conv2D, BatchNormalization, Activation, GlobalAveragePooling2D, Dense, Dropout
from tensorflow.keras.regularizers import l2
from tensorflow.keras.optimizers import Adam

x = base_model.output

x = Conv2D(64, (3,3), padding='same', kernel_regularizer=l2(1e-4))(x)
x = BatchNormalization()(x)
x = Activation('relu')(x)
x = Dropout(0.35)(x)


x = GlobalAveragePooling2D()(x)


x = Dense(128, activation='relu', kernel_regularizer=l2(1e-4))(x)
x = BatchNormalization()(x)
x = Dropout(0.45)(x)

output = Dense(train_data.num_classes, activation='softmax')(x)

model = Model(inputs=base_model.input, outputs=output)

model.compile(
    optimizer=Adam(learning_rate=1e-4),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
model.summary()
```

```python
history = model.fit(
    train_data,
    validation_data=val_data,
    epochs=8
)
```

```python
base_model.trainable = True
for layer in base_model.layers[:-30]:
    layer.trainable = False

model.compile(
    optimizer=Adam(1e-5),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

history_fine = model.fit(
    train_data,
    validation_data=val_data,
    epochs=10
)
```

```python

embedding_model = Model(
    inputs=model.input,
    outputs=model.layers[-3].output
)
```

```python

import numpy as np
from tqdm import tqdm
from tensorflow.keras.preprocessing import image

image_paths = []
embeddings = []

for class_name in os.listdir("dataset"):
    class_folder = os.path.join("dataset", class_name)
    if not os.path.isdir(class_folder):
        continue

    for img_name in os.listdir(class_folder):
        img_path = os.path.join(class_folder, img_name)
        image_paths.append(img_path)

        img = image.load_img(img_path, target_size=(224,224))
        img = image.img_to_array(img) / 255.0
        img = np.expand_dims(img, axis=0)

        emb = embedding_model.predict(img)
        emb = emb.flatten()

        embeddings.append(emb)

embeddings = np.array(embeddings)

print("Embeddings shape:", embeddings.shape)
```

```python
import numpy as np
import pickle

# save embeddings
np.save("embeddings.npy", embeddings)

# save image paths
with open("image_paths.pkl", "wb") as f:
    pickle.dump(image_paths, f)

print("Saved: embeddings.npy and image_paths.pkl")
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
def show_images(images, title="Recommended"):
    plt.figure(figsize=(15, 6))
    for i, img_path in enumerate(images):
        img = mpimg.imread(img_path)
        plt.subplot(1, len(images), i + 1)
        plt.imshow(img)
        plt.axis("off")
        plt.title(f"{title} {i+1}")
    plt.show()
```

```python
def recommend_similar_images(query_image_path, top_k=5):

    img = image.load_img(query_image_path, target_size=(224, 224))
    img = image.img_to_array(img) / 255.0
    img = np.expand_dims(img, axis=0)

    query_emb = embedding_model.predict(img)
    query_emb = query_emb.flatten().reshape(1, -1)

    similarity_scores = cosine_similarity(query_emb, embeddings)[0]

    top_indices = np.argsort(similarity_scores)[::-1][1 : top_k + 1]

    recommended_images = [image_paths[i] for i in top_indices]

    show_images(recommended_images, title="Similar")

    return recommended_images
```

```python
sample_image = image_paths[5128]
show_images([sample_image], title="Query")

recommend_similar_images(sample_image, top_k=5)
```

```python
# ===============================
# 1. IMPORT LIBRARIES
# ===============================
from flask import Flask, request, render_template
import numpy as np
import pickle
from tensorflow.keras.models import load_model, Model
from tensorflow.keras.preprocessing import image
from sklearn.metrics.pairwise import cosine_similarity
from pyngrok import ngrok
import os

# ===============================
# 2. PATHS
# ===============================
BASE_PATH = "/content/drive/MyDrive/boutique"
TEMPLATE_PATH = f"{BASE_PATH}/templates"
STATIC_PATH = f"{BASE_PATH}/static"

# ===============================
# 3. FLASK APP
# ===============================
app = Flask(
    __name__,
    template_folder=TEMPLATE_PATH,
    static_folder=STATIC_PATH
)

# ===============================
# 4. LOAD MODEL & DATA  (UPDATED PART)
# ===============================

# 🔹 Load FULL trained classification model
full_model = load_model(f"{BASE_PATH}/women_boutique_final_model.h5")

# 🔹 Extract Dense(128) embedding layer for QUERY images
model = Model(
    inputs=full_model.input,
    outputs=full_model.layers[-3].output   # Dense(128)
)

# 🔹 Load stored embeddings (already correct)
embeddings = np.load(f"{BASE_PATH}/embeddings.npy")

with open(f"{BASE_PATH}/image_paths.pkl", "rb") as f:
    image_paths = pickle.load(f)

# convert absolute paths → relative paths
image_paths = [
    p.replace("/content/drive/MyDrive/boutique/", "")
    for p in image_paths
]

IMG_SIZE = (224, 224)

# ===============================
# 5. PREPROCESS FUNCTION
# ===============================
def preprocess(img_path):
    img = image.load_img(img_path, target_size=IMG_SIZE)
    img = image.img_to_array(img) / 255.0
    img = np.expand_dims(img, axis=0)
    return img

# ===============================
# 6. HOME ROUTE
# ===============================
@app.route("/", methods=["GET"])
def home():
    return render_template("index.html")

# ===============================
# 7. RECOMMEND ROUTE
# ===============================
@app.route("/recommend", methods=["POST"])
def recommend():
    try:
        if "image" not in request.files:
            return render_template("index.html", images=[])

        file = request.files["image"]
        query_path = "query.jpg"
        file.save(query_path)

        # preprocess
        img = preprocess(query_path)

        # 🔹 QUERY EMBEDDING (NOW 128-DIM)
        query_emb = model.predict(img)
        query_emb = query_emb.flatten().reshape(1, -1)

        # similarity
        sim = cosine_similarity(query_emb, embeddings)[0]
        top_idx = np.argsort(sim)[::-1][1:6]

        # build image URLs
        image_urls = []
        for i in top_idx:
            rel_path = image_paths[i]
            image_urls.append(request.host_url + "static/" + rel_path)

        return render_template("index.html", images=image_urls)

    except Exception as e:
        print("❌ ERROR:", e)
        return f"Internal Error: {e}"

# ===============================
# 8. RUN NGROK + FLASK
# ===============================
public_url = ngrok.connect(5000)
print("OPEN THIS URL:", public_url)

app.run(host="0.0.0.0", port=5000)
```
